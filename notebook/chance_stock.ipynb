## 投机分析

### 个股上榜统计

```python
import tushare as ts
# code：代码
# name:名称
# count：上榜次数
# bamount：累积购买额(万)
# samount：累积卖出额(万)
# net：净额(万)
# bcount：买入席位数
# scount：卖出席位数
#默认为5日
df_cap = ts.cap_tops(days=5)
```

```python
len(df_cap)
```

### 净额买入个股

```python
df_cap_buy = df_cap[(df_cap.net > 0)]
len(df_cap_buy)
```

### 过滤count大于1的个股

```python
df_cap_buy = df_cap_buy[df_cap_buy['count'] > 1]
len(df_cap_buy)
```

```python
#按count排序
df_cap_buy = df_cap_buy.sort_values('count',ascending=False)
df_cap_buy
```

### 净额卖出个股

```python
df_cap_sale = df_cap[df_cap.net < 0]
len(df_cap_sale)
```
